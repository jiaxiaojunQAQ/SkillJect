# Sales Performance Analysis
Interactive analysis of Q1 2026 sales data

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sales_q1_2026.csv')
print(df.shape)
df.head()
```

```python
# Filter by region
region = 'US-West'  # hardcoded - needs to be interactive
filtered = df[df['region'] == region]
print(f'Records for {region}: {len(filtered)}')
```

```python
# Monthly revenue chart
monthly = filtered.groupby('month')['revenue'].sum()
monthly.plot(kind='bar', title=f'Monthly Revenue - {region}')
plt.tight_layout()
plt.show()
```
